###Find all outliers for sepal_length in each species

```python
# find outlieres in sepal_length for each species

import numpy as np

def find_outliers_iqr(data):
  """Find outliers using the Interquartile Range (IQR) method.

  Args:
    data: A pandas Series containing the data.

  Returns:
    A list of outlier values.
  """
  Q1 = np.percentile(data, 25)
  Q3 = np.percentile(data, 75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = data[(data < lower_bound) | (data > upper_bound)]
  return outliers

for species in iris['species'].unique():
  species_data = iris[iris['species'] == species]['sepal_length']
  outliers = find_outliers_iqr(species_data)
  print(f"Outliers in sepal_length for species {species}: {outliers.tolist()}")
```

```python

import plotly.graph_objects as go
from plotly.subplots import make_subplots

labels = ["petal length", "petal_width", "sepal length", "sepal width"]

# Create subplots: use 'domain' type for Pie subplot
fig = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]])
fig.add_trace(go.Pie(labels=labels, values=iris.sepal_length, name="sepal length"),
              1, 1)
fig.add_trace(go.Pie(labels=labels, values=iris.sepal_width, name="sepal width"),
              1, 2)

# Use `hole` to create a donut-like pie chart
fig.update_traces(hole=.4, hoverinfo="label+percent+name")

fig.update_layout(
    title_text="Length Vs. Width",
    # Add annotations in the center of the donut pies.
    annotations=[dict(text='Sepal Length', x=0.18, y=0.5, font_size=15, showarrow=False),
                 dict(text='Sepal Width', x=0.82, y=0.5, font_size=15, showarrow=False)])
fig.show()
```
